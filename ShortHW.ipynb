```python
# RandomState: controls the shuffling applied to the data before the split. 
rs = np.random.RandomState(0)
# in general, %85 of your data for training and 25% of data for test.
# choise test index sets from data
test = rs.choice(irisCleaned.index, len(irisCleaned) // 4)
train = irisCleaned.index[~irisCleaned.index.isin(test)]

print(test)
print(train)
```

```python
irisCleaned.loc[train, 'subset'] = 'Train'
irisCleaned.loc[test, 'subset'] = 'Test'
irisTrain = irisCleaned[irisCleaned['subset'] == 'Train']
irisTest = irisCleaned[irisCleaned['subset'] == 'Test']
irisCleaned.groupby('subset').count()[['Species']]
```

```python
#bias hesaplama fonksiyonu
def addBiasTerm(x):
    # Return a new array of given shape and type, filled with ones.
    b = np.ones((len(x), 1)) 
    ## Stack arrays in sequence horizontally (column wise).
    x = np.hstack([b, x]) 
    return x
```

```python
xTrain = irisTrain[['PetalLengthCm']].values
xTrain = addBiasTerm(xTrain)
yTrain = irisTrain[['PetalWidthCm']].values
xTrain.shape, yTrain.shape
```

```python
xTrain[:3], yTrain[:3]
```

```python
# pinv(): pseudo-inverse of a matrix
beta = lambda x, y: np.linalg.pinv(x.T @ x) @ x.T @ y
betaTrain = beta(xTrain, yTrain)
betaTrain
```

```python
# .transpose: Reverse or permute the axes of an array; returns the modified array
RSS = lambda x, y, beta: np.transpose(y - x @ beta) @ (y - x @ beta)
RSS(xTrain, yTrain, betaTrain)
```

```python
MSE = lambda x, y, b: (RSS(x, y, b) / len(x)).item()
errTrain = MSE(xTrain, yTrain, betaTrain)
errTrain
```

```python
xTest = irisTest[['PetalLengthCm']].values
xTest = addBiasTerm(xTest)
yTest = irisTest[['PetalWidthCm']].values
xTest.shape, yTest.shape
```

```python
xTest[:3], yTest[:3]
```

```python
errTest = MSE(xTest, yTest, betaTrain)
errTest
```

```python
#predict yapmka için gerekli fonksiyonun tanımı
def predict(xAxis, beta):
    x = np.linspace(xAxis.min(), xAxis.max())
    b = np.ones(len(x))
    # Stack arrays in sequence vertically (row wise).
    X = np.vstack([b, x]).T
    yHat = X @ beta
    yHat = yHat.flatten()
    return pd.DataFrame({'x': x, 'y': yHat})
```

```python
yHat = predict(irisCleaned['PetalLengthCm'], betaTrain)
yHat.shape
```
